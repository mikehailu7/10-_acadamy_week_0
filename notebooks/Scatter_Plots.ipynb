```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('./data/benin-malanville.csv')
```

```python
scatter_pairs_1 = [('GHI', 'Tamb'), ('WS', 'WSgust')]
```

```python
plt.figure(figsize=(12, 6))
for i, pair in enumerate(scatter_pairs_1, start=1):
    plt.subplot(1, len(scatter_pairs_1), i)
    plt.scatter(data_1[pair[0]], data_1[pair[1]], alpha=0.5)
    plt.title(f'{pair[0]} vs. {pair[1]}')
    plt.xlabel(pair[0])
    plt.ylabel(pair[1])

plt.tight_layout()
```
