```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_1 = pd.read_csv('./data/benin-malanville.csv')
```

```python
data_1['Timestamp'] = pd.to_datetime(data_1['Timestamp'])
```

```python
data_1.set_index('Timestamp', inplace=True)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(data_1['WS'], color='blue', label='WS')
plt.plot(data_1['WSgust'], color='green', label='WSgust')
plt.plot(data_1['WSstdev'], color='red', label='WSstdev')
plt.title('Wind Speed')
plt.xlabel('Timestamp')
plt.ylabel('Wind Speed (m/s)')
plt.legend()
```

```python
plt.subplot(2, 2, 2)
plt.plot(data_1['WD'], color='blue', label='WD')
plt.plot(data_1['WDstdev'], color='green', label='WDstdev')
plt.title('Wind Direction')
plt.xlabel('Timestamp')
plt.ylabel('Wind Direction (degrees)')
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_1['WS'], bins=20, kde=True, color='blue')
plt.title('Histogram of Wind Speed')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.show()
```

```python
from windrose import WindroseAxes
```

```python
plt.figure(figsize=(10, 10))
ax = WindroseAxes.from_ax()
ax.bar(data_1['WD'], data_1['WS'], normed=True, opening=0.8, edgecolor='white')
ax.set_legend(title='Wind Speed (m/s)')
plt.title('Wind Rose Plot')
plt.show()
```
