```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('./data/benin-malanville.csv')
```

```python
data_1['Timestamp'] = pd.to_datetime(data_1['Timestamp'])
```

```python
data_1.set_index('Timestamp', inplace=True)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(data_1['GHI'], color='blue')
plt.title('Global Horizontal Irradiance (GHI)')
plt.xlabel('Timestamp')
plt.ylabel('GHI')
```

```python
plt.subplot(2, 2, 2)
plt.plot(data_1['DNI'], color='red')
plt.title('Direct Normal Irradiance (DNI)')
plt.xlabel('Timestamp')
plt.ylabel('DNI')
```

```python
plt.subplot(2, 2, 3)
plt.plot(data_1['DHI'], color='green')
plt.title('Diffuse Horizontal Irradiance (DHI)')
plt.xlabel('Timestamp')
plt.ylabel('DHI')
```

```python
plt.subplot(2, 2, 4)
plt.plot(data_1['Tamb'], color='orange')
plt.title('Ambient Temperature (Tamb)')
plt.xlabel('Timestamp')
plt.ylabel('Tamb')
```

```python
plt.tight_layout()
plt.show()
```
