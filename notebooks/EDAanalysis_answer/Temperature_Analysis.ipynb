```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('./data/benin-malanville.csv')
```

```python
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 2, 1)
plt.scatter(data_1['Tamb'], data_1['TModA'], color='blue', alpha=0.5)
plt.title('TModA vs Tamb')
plt.xlabel('Ambient Temperature (Tamb)')
plt.ylabel('Module Temperature A (TModA)')
```

```python
plt.subplot(1, 2, 2)
plt.scatter(data_1['Tamb'], data_1['TModB'], color='red', alpha=0.5)
plt.title('TModB vs Tamb')
plt.xlabel('Ambient Temperature (Tamb)')
plt.ylabel('Module Temperature B (TModB)')

plt.tight_layout()
plt.show()
```

```python
correlation_TModA_1 = data_1['Tamb'].corr(data_1['TModA'])
correlation_TModB_1 = data_1['Tamb'].corr(data_1['TModB'])

print("Correlation coefficient between Tamb and TModA:", correlation_TModA_1)
print("Correlation coefficient between Tamb and TModB:", correlation_TModB_1)
```
