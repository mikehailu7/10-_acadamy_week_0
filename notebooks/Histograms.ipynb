```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('./data/benin-malanville.csv')
```

```python
plt.figure(figsize=(12, 10))
```

```python
plt.subplot(3, 2, 1)
plt.hist(data_1['GHI'], bins=20, color='blue', alpha=0.7)
plt.title('Histogram of GHI')
plt.xlabel('GHI')
plt.ylabel('Frequency')
```

```python
plt.subplot(3, 2, 2)
plt.hist(data_1['DNI'], bins=20, color='green', alpha=0.7)
plt.title('Histogram of DNI')
plt.xlabel('DNI')
plt.ylabel('Frequency')
```

```python
plt.subplot(3, 2, 3)
plt.hist(data_1['DHI'], bins=20, color='orange', alpha=0.7)
plt.title('Histogram of DHI')
plt.xlabel('DHI')
plt.ylabel('Frequency')
```

```python
plt.subplot(3, 2, 4)
plt.hist(data_1['WS'], bins=20, color='red', alpha=0.7)
plt.title('Histogram of WS (Wind Speed)')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
```

```python
plt.subplot(3, 2, 5)
plt.hist(data_1['Tamb'], bins=20, color='purple', alpha=0.7)
plt.title('Histogram of Tamb (Ambient Temperature)')
plt.xlabel('Ambient Temperature (Tamb)')
plt.ylabel('Frequency')
```

```python
plt.subplot(3, 2, 6)
plt.hist(data_1['TModA'], bins=20, color='brown', alpha=0.7)
plt.title('Histogram of TModA (Module Temperature A)')
plt.xlabel('Module Temperature A (TModA)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
