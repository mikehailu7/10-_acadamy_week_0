```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('./data/benin-malanville.csv')
```

```python
columns_of_interest_1 = ['GHI', 'DNI', 'DHI', 'Tamb', 'TModA', 'TModB']
```

```python
plt.figure(figsize=(10, 6))
data_1[columns_of_interest_1].boxplot()
plt.title('Box Plot of Solar Radiation and Temperature Data')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```
