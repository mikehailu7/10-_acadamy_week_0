```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv('./data/benin-malanville.csv')
```

```python
missing_value_1 = data_1.isnull().sum()
```

```python
print("Missing Values:\n", missing_value_1)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data_1=data_1[['GHI', 'DNI', 'DHI']])
plt.title('Boxplot of GHI, DNI, and DHI')
plt.show()
```

```python
negative_value_1 = data_1[(data_1['GHI'] < 0) | (data_1['DNI'] < 0) | (data_1['DHI'] < 0)]
print("Entries with negative values:\n", negative_value_1)
```
